```python
import requests
from bs4 import BeautifulSoup

page = requests.get('https://www.epfl.ch/labs/')

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
acronyms = soup.select("div[class*=col-2]")[1:]
```

```python
lab_names = soup.select("div[class*=col-7]")[1:]
```

```python
faculty = soup.select("a[class*=\"tag tag-primary site-tags-faculty\"]")
```

```python
institute = soup.select("a[class*=\"tag tag-primary site-tags-institute\"]")
```

```python
def extract_string(s, leader, trailer):
    end_of_leader = s.index(leader) + len(leader)
    start_of_trailer = s.index(trailer, end_of_leader)
    return s[end_of_leader:start_of_trailer]
```

```python
entries = soup.select("div[class=\"site-row flex-row d-md-flex pt-1 pb-1 border-bottom align-items-center]\"")
```

```python
entries[0]
```

```python
for entry in entries:
    entry = str(entry)
    acronym = extract_string(entry, "<div class=\"site-title col-2\">", "</div>")
    source = extract_string(entry, "<div class=\"col-7\"><a class=\"site-url\" href=\"", "\"><span")
    name = extract_string(entry, "<span class=\"site-tagline\">", "</span></a></div>")
    if "faculty" in entry:
        faculty = extract_string(entry[entry.index("faculty"):], ">", "</a>")
    else:
        faculty = ""
    if "institute" in entry:
        institute = extract_string(entry[entry.index("institute"):], ">", "</a>")
    else:
        institute = ""
        
    line = name + "," + acronym + "," + source + "," + faculty + "," + institute + ",,"
    print(line)
```
